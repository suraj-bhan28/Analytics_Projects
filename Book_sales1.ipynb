```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import csv file
df=pd.read_csv('Books_Data_Clean.csv',encoding='unicode_escape')
```

```python
df.head()
```

```python
df.describe()
```

```python
df=df[df["Publishing Year"] >1900]
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.dropna(subset="Book Name",inplace=True)
```

```python
df.isna().sum()
```

```python
df.duplicated()
```

```python
df.nunique()
```

```python
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Ditribution of Publishing Year")
plt.show()
```

```python
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("genre")
plt.ylabel("Number of books")
plt.title("Number of Bokks is each genre")
plt.show()
```

```python
df.groupby("Author")["Book_average_rating"].mean().head(5)
```

```python
sns.boxplot(x="genre",y="Book_ratings_count",data=df)
plt.xlabel("genre")
plt.ylabel("Book Rating Count")
plt.title("Number of book rating count for each genre")
plt.show()
```

```python
plt.scatter(df["sale price"],df["units sold"])
plt.xlabel("sale price")
plt.ylabel("units sold")
plt.title("Scatter plot of sale Price vs unoits sold")
plt.show()
```

```python
language_count=df["language_code"].value_counts().sort_values(ascending =True)
```

```python
plt.pie(language_count,labels=language_count.index,startangle=90,autopct="%1.1f%%")
plt.title("Language Distribution of Books")
plt.show()
```

```python
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False).max()
```

```python
plt.scatter(df["Book_average_rating"],df["Book_ratings_count"])
plt.xlabel("Book_average_rating")
plt.ylabel("Book_ratings_count")
plt.title("Scatter plot of Book avrrage rating count  vs Book_ratings_count")
plt.show()
```

```python
Total_gross_sales_by_author=df.groupby("Author")["gross sales"].sum()
```

```python
Total_gross_sales_by_author.sort_values(ascending=False).head(5).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total gross sales for each author")
plt.show()
```
